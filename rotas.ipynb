```python
df = pd.read_excel('Pasta1.xlsx', sheet_name='Planilha1')
df.head()
```

```python
print(df.columns.tolist())
```

```python
rotas_existentes = df[['Unnamed: 1', 'Unnamed: 2']].drop_duplicates().reset_index(drop=True)
print(rotas_existentes)
```

```python
# Criar a matriz cruzada (quantas vezes cada fábrica atendeu cada cliente)
matriz_rotas = pd.crosstab(df['Unnamed: 1'], df['Unnamed: 2'])
matriz_rotas
```

```python
import pandas as pd

df_nomes = pd.read_excel('Pasta2.xlsx')

df_nomes.head()
```

```python
# Separando dicionários
dic_fabricas = df_nomes.set_index('Unnamed: 3')['Unnamed: 4'].to_dict()
dic_clientes = df_nomes.set_index('Unnamed: 1')['Unnamed: 2'].to_dict()
```

```python
# Renomeia índices (linhas) com nomes das fábricas
matriz_rotas.rename(index=dic_fabricas, inplace=True)

# Renomeia colunas com nomes dos clientes
matriz_rotas.rename(columns=dic_clientes, inplace=True)
```

```python
matriz_rotas.head()
```

```python
matriz_rotas.to_excel('matriz_rotas_nomeada.xlsx')
```

```python
from google.colab import files
files.download('matriz_rotas_nomeada.xlsx')
```

```python
# Converte a matriz em formato de lista
matriz_lista = matriz_rotas.reset_index().melt(id_vars=matriz_rotas.index.name,
                                               var_name='Cliente',
                                               value_name='Quantidade de Rotas')

# Opcional: remover linhas com 0 rotas
matriz_lista = matriz_lista[matriz_lista['Quantidade de Rotas'] > 0]
```
